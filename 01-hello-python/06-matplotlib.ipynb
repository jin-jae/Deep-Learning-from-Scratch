# Draw Graph

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 6, 0.1) # zero to six, 0.1 steps
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

# pyplot

```python
x = np.arange(0, 6, 0.1) # zero to six, 0.1 steps
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # -- type line
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend() # show legend
plt.show()
```
